### Q: Write a Pandas program to create and display a DataFrame from the following dictionary data which has index labels.
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

```python
import pandas as pd
import numpy as np

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

df= pd.DataFrame(exam_data, columns=['name','score','attempts','qualify'], index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])

df
```

### 1:Print the three first rows using head() method

```python
df.head(3)
```

### 2:Delete rows with Nan values

```python
df.dropna(inplace=True)
df
```

### 3.Extract the 'name' and 'score' columns from the DataFrame.

```python
df[['name', 'score']]
```

### 4.Write a Pandas program to append a new row 'k' to data frame with these values(name : "Suresh", score: 15.5, attempts: 1, qualify: "yes")

```python
d1={'name' : "Suresh", 'score': 15.5, 'attempts': 1, 'qualify': "yes"}
df1=pd.DataFrame(d1, columns=['name','score','attempts','qualify'], index=['k'])
df=pd.concat([df,df1])
df
```

### 5.Write a Pandas program to delete the 'attempts' column from the DataFrame.

```python
df.drop("attempts",axis=1,inplace=True)
df
```

### 6.Add a new column "Success" : if the score is higher than 10 we will have 1 else we will have 0

```python
df['Success'] = [1 if x > 10 else 0 for x in df['score']]
```
